```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["font.family"] = "Times New Roman"

# Models
models = ["L-14", "S4", "S3", "B", "S2", "S0"]  # reversed so top = L-14
y = np.arange(len(models))

# Scores (%)
k0 = np.array([0.8372, 0.8509, 0.833, 0.826, 0.804, 0.749]) * 100      # zero-shot
k_all = np.array([0.997, 0.990, 0.988, 0.984, 0.9805, 0.9666]) * 100    # trained on all data

fig, ax = plt.subplots(figsize=(7, 4))
ax.set_facecolor("#efefef")

# Dumbbell lines
for i in range(len(models)):
    ax.plot([k0[i], k_all[i]], [y[i], y[i]],
            color="gray", linewidth=1.2, zorder=1)

# Zero-shot stars
ax.scatter(k0, y, marker="*", s=150, color="purple",
           label="Zero-shot (K=0)", zorder=3)

# Trained-on-all dots
ax.scatter(k_all, y, marker="o", s=70, color="C0",
           label="+ CoOp (K=10)", zorder=3)

# Axes
ax.set_yticks(y)
ax.set_yticklabels(models)
ax.set_xlabel("Accuracy (%)")
ax.set_title("Flower102: Zero-shot vs CoOp (K=10)")

ax.grid(axis="x", color="white", linewidth=1.2)
ax.set_axisbelow(True)
ax.set_xlim(72, 100)

# --- Legend: Move left + add box ---
legend = ax.legend(
    loc="center left",
    bbox_to_anchor=(0.03, 0.2),  # move further left if needed
    frameon=True,                 # show box
    framealpha=1.0,               # solid box
    edgecolor="black",            # black outline
    facecolor="white"             # white legend background
)

plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["font.family"] = "Times New Roman"

# Models
models = ["0", "1", "2", "4", "8", "all=10"]  
y = np.arange(len(models))

# Scores (%)
mobileclip = np.array([0.7521, 0.8235, 0.8728, 0.9208, 0.946, 0.963]) * 100      # MobileCLIP2
vit = np.array([0.714, 0.783, 0.88, 0.922, 0.949, 0.968]) * 100    
# transclip = np.array([0.707, 0.848, 0, 0.936, 0, 0.944]) * 100

fig, ax = plt.subplots(figsize=(7, 4))
ax.set_facecolor("#efefef")

# Dumbbell lines
for i in range(len(models)):
    p1 = min(mobileclip[i], vit[i])
    # if p1 == 0:
    #     p1 = min(mobileclip[i], vit[i])
    p2 = max(mobileclip[i], vit[i])
    ax.plot([p1, p2], [y[i], y[i]],
            color="gray", linewidth=1.2, zorder=1)

# Zero-shot stars
ax.scatter(mobileclip, y, marker=r"$M$", s=150, color="#EF7000",
           label="MobileCLIP-S0", zorder=3)

# Trained-on-all dots
ax.scatter(vit, y, marker="o", s=70, color="C0",
           label="ViT-B/16 CLIP*", zorder=3)

# ax.scatter(transclip, y, marker="v", s=70, color="green",
#            label="TransCLIP", zorder=3)

# Axes
ax.set_yticks(y)
ax.set_yticklabels(models)
ax.set_xlabel("Accuracy (%)")
ax.set_ylabel("Number of shots")
ax.set_title("Flower102: MobileCLIP-S0 vs ViT-B/16")

ax.grid(axis="x", color="white", linewidth=1.2)
ax.set_axisbelow(True)
ax.set_xlim(70, 100)

# --- Legend: Move left + add box ---
legend = ax.legend(
    loc="center left",
    bbox_to_anchor=(0.03, 0.85),  # move further left if needed
    frameon=True,                 # show box
    framealpha=1.0,               # solid box
    edgecolor="black",            # black outline
    facecolor="white"             # white legend background
)

plt.tight_layout()
plt.show()
```

```python


import matplotlib.pyplot as plt
import numpy as np

# Labels
labels = ["A", "A photo of a", "A flower of type", "A picture of the flower"]

# Matrix values
matrix = np.array([
    [1,     0.748, 0.708, 0.639],
    [0.748, 1,     0.658, 0.584],
    [0.708, 0.658, 1,     0.895],
    [0.639, 0.584, 0.895, 1]
])

plt.figure(figsize=(5,4))

# Plot heatmap
plt.imshow(matrix)

# Add colorbar
plt.colorbar()

# Label ticks
plt.xticks(range(len(labels)), labels, rotation=-90, ha='right', fontsize=12)
plt.yticks(range(len(labels)), labels, fontsize=12)

# Add numbers inside each cell
for i in range(matrix.shape[0]):
    for j in range(matrix.shape[1]):
        plt.text(j, i, f"{matrix[i, j]:.3f}",
                 ha='center', va='center', color='white' if matrix[i,j] < 0.75 else 'black',
                 fontsize=12)

plt.title("Cosine Similarity of Handcrafted Prompts on Flower102", pad=20)
plt.tight_layout()
plt.show()
```
